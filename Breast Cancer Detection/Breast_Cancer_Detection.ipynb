# Defining Features and labels from the dataset

-> Features are like input which is used for further processing of data and it is also given to model.

-> Labels are like output which we expect from the model

```python
features = data['data']
label = data['target']
```

```python
features.shape
```

```python
label.shape
```

```python
scaled_features = StandardScaler()
features = scaled_features.fit_transform(features)
```

# Converting dataset into dataframe

```python
# First of all we have to create Dataframe for features as well as for labels and 
# then merge / Concatenate those to make one Dataframe

df_feat = pd.DataFrame(columns=data['feature_names'], data=features)
df_lab  = pd.DataFrame(columns=['label'], data=label)

df = pd.concat([df_feat, df_lab], axis=1)
df.head()
```

# Splitting Dataset for Training and Testing

```python

#500 Training
X_train = features[:500]
y_train = label[:500]

#35 Validation
X_val = features[500:535]
y_val = label[500:535]

#34 Testing
X_test = features[535:]
y_test = label[535:]
```

# Model Building

```python
model = Sequential()

model.add(Dense(32, activation = 'relu', input_dim = 30))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile( loss = 'binary_crossentropy' , optimizer = 'adam' , metrics = ['accuracy'])
```

```python
model.fit( X_train , y_train, epochs = 10, batch_size = 5, validation_data = (X_val, y_val))
```

```python
model.evaluate(X_test , y_test)
```

```python
model.evaluate(X_val , y_val)
```

```python
for i in range(34):
  sample = X_test[i]
  sample = np.reshape(sample, (1,30))

  if (model.predict(sample)[0][0] > 0.5):
    print("-Benign")
  else:
    print("-Malignant")

  if (y_test[i] == 1):
    print("*Banign")
  else:
    print("*Melignant")
  print("-----------")
```
